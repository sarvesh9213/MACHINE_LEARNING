# Steps of EDA

- There are no specific steps for EDA.
- It depends upon the problem or what insights we want to draw from the data.
- Still there are a few steps that we follow with any kind of data that we are handling.
- After following those steps we can have some good understanding about the data.

```python
# - We check the size of the data (Shape of the data)
# - We check the attributes (columns) in the data.
# - We check the 5 point statistical summary of the data.
# - We check for the null values in the data
# - Treat the null values if present
# - We check for the duplicates
# - Treat the duplicate values if present
# - We check for the outliers
# - Treat the outliers if present
# - We check the data distribution
# - Analysis (Depending upon the insights we want to draw)
```

# Imputation of the missing values

```python
# Generally in a Numeric continuous column we fill the missing values with mean or median
# In categorical column we fill with mode
```

```python
data.nunique()
```

```python
num_cols=[]
cat_cols=[]

for i in data.columns:
    if data[i].nunique()>5 and data[i].dtype==float or data[i].dtype==int:
        num_cols.append(i)
    else:
        cat_cols.append(i)
```

```python
num_cols
```

```python
cat_cols
```

```python
def null_impute_numeric(col):
    data[col].fillna(data[col].mean(),inplace=True)
```

```python
def null_impute_cat(col):
    data[col].fillna(data[col].mode()[0],inplace=True)
```

```python
for i in num_cols:
    null_impute_numeric(i)
```

```python
data.isna().sum()
```

```python
for i in cat_cols:
    null_impute_cat(i)
```

```python
data.isna().sum()
```

# Treatment of the outliers¶

```python
num_cols.remove('Inquiries_in_the_Last_6_Months')
num_cols.remove('Employment_Length')
```

```python
num_cols
```

```python
q3=data.Amount_Requested.quantile(.75)
q3
```

```python
q1=data.Amount_Requested.quantile(.25)
q1
```

```python
iqr=q3-q1
iqr
```

```python
upper=q3+1.5*iqr
upper
```

```python
lower=q1-1.5*iqr
lower
```

```python
data.Amount_Requested.clip(lower,upper,inplace=True)
```

```python
data.Amount_Requested.describe()
```

```python
# Try using Z score approach, UL = mean+ 3SD, LL = mean - 3SD
```

```python
# Outlier capping
def out_cap(col):
    q3 = data[col].quantile(0.75)
    q1 = data[col].quantile(0.25)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    data[col] = data[col].clip(lower, upper)

```

```python
for col in num_cols:
    out_cap(col)
```

```python
# Visualization
fig, ax = plt.subplots(2, 4, figsize=(18, 15), dpi=100)
ax = ax.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(x=data[col], ax=ax[i])
    ax[i].set_title(col, pad=8)

for j in range(len(num_cols), len(ax)):
    ax[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
s=data.groupby('Loan_Purpose').Amount_Requested.sum().astype(int).div(1000).round(2)
s_sorted = s.sort_values(ascending=False)
s_sorted
```

```python
plt.figure(figsize=(18,8),dpi=100)

a=sns.barplot(x=s_sorted.index,y=s_sorted.values)

for i in a.containers:
    a.bar_label(i,fmt='%d')
    
plt.xticks(rotation=45)

plt.show()
```
