## Missing values and imputing them

```python
# find out the percentage of missing values
df.isnull().sum() / len(df) * 100
```

Pandas doesn't know:

Is "Age" a row label?
Or is "Age" a column name?

So axis removes the ambiguity.

axis=0 → Look in the row index.
axis=1 → Look in the column names.

```python
#1- drop the column having more than 70% of missing values from dataset
df.drop('deck', axis=1) # isse actual data mai changes nhi aa paate hai 
# also iske ander dekho toh petaa chlega that [] is not used 
df.drop('deck', axis=1, inplace=True)
```

```python
df.columns # isse ab deck gayab ho gaya hai because of inplace equals to true 
```

```python
df.isnull().sum()
```

```python
df['age'].mean()
```

```python
df['age'].median() # numceric value ko fill kerne ke liyee kerte hai
```

```python
df['age'].mode()[0] # zayda ter cateogrical ke liye use kerte hai 
```

```python
# 2- filling missing values of age column by the mean of age
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
# df['age'].fillna(df['age'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# fill embark and embarked_town with mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

# IN SIMPLE HOW TO HANDLE THE MISSING VALUES

```python
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
sns.heatmap(df.isnull())
```
