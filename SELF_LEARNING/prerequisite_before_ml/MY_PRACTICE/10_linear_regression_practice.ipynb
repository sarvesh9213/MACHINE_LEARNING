# IMPORT LIBRARIES

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib as plt 
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error,r2_score
```

# LOAD DATA-SET

```python
df=sns.load_dataset('tips')
df
```

```python
df.isnull().sum()
```

```python
# by sir !!! 
sns.scatterplot(x='total_bill',y='tip',data=df)
```

# SELECTING X AND Y

### TOTAL BILL -> MODEL -> PREDICTED TIP

```python
X = df[['total_bill']]
y = df['tip']
# Scikit-learn expects X to be 2-dimensional
# [[10.0],
#  [20.0],
#  [30.0]]
```

# APPLY-> { TRAIN - TEST DATA SPLIT }

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
# also phele train test x then train test y
# yaha per mai train and test dono daal rha tha bracket mai and which is not correct 
```

# CREATE THE MODEL

```python
model = LinearRegression()
```

# LET'S TRAIN THE MODEL 

```python
model.fit(x_train,y_train)
```

```python
# finds the best possible line:
# ^y=mx+b
# Scikit-learn automatically calculates the best m and b.
```

```python
print(model.coef_)
print(model.intercept_)
```

# Make predictions

```python
y_prediction = model.predict(x_test)
```

```python
# model.fit(x_train,y_train)
# model.coef_
# model.intercept_
# model.predict(x_test)
```

```python
# y_test = Actual tips
# y_pred = Model's predicted tips
```

# EVALUATE THE MODEL :   MSE/RMSE/R2

```python
rmse = root_mean_squared_error(y_test,y_prediction)
```

```python
print(rmse)
```

```python
r2_score = r2_score(y_test,y_prediction)
```

```python
print(r2_score)
```

```python
print("coefficient",model.coef_[0])
```

```python
print("intercept",model.intercept_)
```

# VISUALIZE THE MODEL

```python
import matplotlib.pyplot as plt 
plt.scatter(x_test, y_test)
plt.plot(x_test, y_prediction , color = 'red')

plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```
